#SELECT EXPERIMENT FOLDER

Prompt the user to select the experiment root directory via a GUI dialog.
The returned path is passed into the main scoring pipeline.


```python
from google.colab import drive
from IPython.display import display
import ipywidgets as widgets
from ipyfilechooser import FileChooser
import os
import sys

drive.mount('/content/drive')


# Add the folder where your Python modules are stored to the Python path
sys.path.append('/content/drive/MyDrive/Matheus_e_Rodrigo/Refactored_codes/')  # Find the folder
sys.path.append('/content/drive/MyDrive/Matheus_e_Rodrigo/Refactored_codes/BehaviorScoringCodes/')
sys.path.append('/content/drive/MyDrive/Matheus_e_Rodrigo/Refactored_codes/BehaviorScoringCodes/BehaviorScoring/')
import ExperimentConfig as EConfig
import BehaviorScoringFunctions_V0_0_1 as BSF
import BehaviorScoringMain_V0_0_2____RESISTANT_ALL as BSM

# Create a widget that allows directory selection
chooser = FileChooser('/content/drive/MyDrive/Matheus_e_Rodrigo/')
chooser.title = 'Select the experiment folder'
chooser.show_only_dirs = True  # Only allow selecting directories
```

#Select Folder

```python
# Display the widget
display(chooser)

```

#BEHAVIOR SCORING

Load experiment configuration and scoring logic, then run the main pipeline.
##Arguments:
    EXPERIMENT_PATH: path to the experiment root (from CELL¬†01)
    EConfig         : constants defining arena, thresholds, filenames, etc.
    BSF             : behaviorscoring helper functions

```python


# Create a run button and an output area
run_button = widgets.Button(description="Run Analysis", button_style='success')
output = widgets.Output()

# Define what happens when the button is clicked
def on_run(b):
    with output:
        output.clear_output()
        exp_path = chooser.selected_path
        if exp_path:
            print(f"Starting pipeline for: {exp_path}")
            BSM.main_pipeline(exp_path, EConfig, BSF)
            print("‚úÖ Analysis completed.")
        else:
            print("‚ö†Ô∏è No folder selected.")

# Connect the button to the function
run_button.on_click(on_run)



```

#Run

```python
# Show the button and output
display(run_button, output)
```
